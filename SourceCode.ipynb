 data preprocessing

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
      #feature engineering
#temperature range
df['Present_TempRange'] = df['Present_Tmax'] - df['Present_Tmin']
df['Next_TempRange'] = df['Next_Tmax'] - df['Next_Tmin']
```

```python
#data-based feature
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.dayofweek  # 0 = Monday
```

```python
#encoding
df['station'] = df['station'].astype('category')
df['station_cat'] = df['station'].cat.codes
```

```python
df.drop(columns=['station'], inplace=True)
```

modeling

```python
# Split features and target
x = df.drop(columns=['Date', 'Next_Tmax', 'Next_Tmin'])
y = df[['Next_Tmax', 'Next_Tmin']]
```

```python
# train the model
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
# fit a model
base_model = RandomForestRegressor(n_estimators=100, random_state=42)
multi_model = MultiOutputRegressor(base_model)

multi_model.fit(X_train, y_train)
```

```python
#predict the model
y_pred = multi_model.predict(X_test)
```

```python
# Split predictions
pred_tmax = y_pred[:, 0]
pred_tmin = y_pred[:, 1]
```

```python
# True values
true_tmax = y_test['Next_Tmax'].values
true_tmin = y_test['Next_Tmin'].values
```

```python
# Evaluation
print(" Next_Tmax Evaluation:")
print(f"MAE: {mean_absolute_error(true_tmax, pred_tmax):.2f} Â°C")
print(f"RMSE: {np.sqrt(mean_squared_error(true_tmax, pred_tmax)):.2f} Â°C")
print(f"RÂ² Score: {r2_score(true_tmax, pred_tmax):.2f}")

print("\n Next_Tmin Evaluation:")
print(f"MAE: {mean_absolute_error(true_tmin, pred_tmin):.2f} Â°C")
print(f"RMSE: {np.sqrt(mean_squared_error(true_tmin, pred_tmin)):.2f} Â°C")
print(f"RÂ² Score: {r2_score(true_tmin, pred_tmin):.2f}")
```

```python
# Feature importance
tmax_importances = pd.Series(multi_model.estimators_[0].feature_importances_, index=x.columns)
tmax_importances.sort_values().plot(kind='barh', figsize=(10, 6), color='blue')
plt.title("Feature Importance for Predicting Next_Tmax")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```

```python
tmin_importances = pd.Series(multi_model.estimators_[1].feature_importances_, index=x.columns)
tmin_importances.sort_values().plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.title("Feature Importance for Predicting Next_Tmin")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```
